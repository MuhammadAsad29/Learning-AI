```python
from collections import deque

# Define the warehouse grid (S = Start, G = Goal, X = Obstacle, . = Open Path) 
warehouse_grid = [
['S', '.', '.', 'X', 'G'],
['.', 'X', '.', '.', '.'],
['.', '.', '.', 'X', '.'],
['X', '.', '.', '.', '.'],
['.', '.', 'X', '.', '.']
]

# Define start and goal positions 
start_position = (0, 0)
goal_position = (0, 4)

# Function to get neighbors in grid 
def get_neighbors(position, grid):
    row, col = position
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)] # Up, Down, Left, Right 
    neighbors = []
    
    for dr, dc in directions:
        r, c = row + dr, col + dc
        if 0 <= r < len(grid) and 0 <= c < len(grid[0]) and grid[r][c] != 'X': 
            neighbors.append((r, c))
    return neighbors
    
# Breadth-First Search (BFS) implementation 
def bfs(start, goal, grid):
    queue = deque([start]) # Queue for BFS
    visited = set([start]) # Track visited cells 
    steps = 0
    while queue:
        position = queue.popleft() 
        if position == goal:
            return steps, len(visited) # Return steps and nodes visited 
        for neighbor in get_neighbors(position, grid):
            if neighbor not in visited: 
                visited.add(neighbor) 
                queue.append(neighbor)
        steps += 1
    return None

# Depth-First Search (DFS) implementation 
def dfs(start, goal, grid):
    stack = [start] # Stack for DFS
    visited = set([start]) # Track visited cells 
    steps = 0

    while stack:
        position = stack.pop() 
        if position == goal:
            return steps, len(visited) # Return steps and nodes visited
        for neighbor in get_neighbors(position, grid): 
            if neighbor not in visited:
                visited.add(neighbor) 
                stack.append(neighbor)
        steps += 1 
    return None
    
# Run BFS and DFS
bfs_steps, bfs_visited = bfs(start_position, goal_position, warehouse_grid) 
dfs_steps, dfs_visited = dfs(start_position, goal_position, warehouse_grid) # Print the results

print("BFS Steps to Goal:", bfs_steps)
print("BFS Nodes Visited:", bfs_visited) 
print("DFS Steps to Goal:", dfs_steps) 
print("DFS Nodes Visited:", dfs_visited)
```
